Proces normowania kolumn

```python
data = pd.read_excel("Dane.xlsx", usecols="G:K")
#otrzymanie wymiarów utworzonej tablicy
numbers = data.shape
#zczytanie liczby kolumn oraz liczby wierszy utworzonej tablicy
number_of_columns = numbers[1]
number_of_rows = numbers[0]
array = data.to_numpy()
#prelokacja tablic, tak aby uniknąć dynamicznego alokowania pamięci - uzupełnienie tablic zerami
maxelement = np.zeros(number_of_columns)
normalized_array = np.zeros((number_of_rows, number_of_columns))
#proces normowania kolumn
for i in range(number_of_columns):
  maxelement[i] = np.amax(array[0:number_of_rows, i])
  for j in range(number_of_rows):
    normalized_array[j, i] = array[j, i]/maxelement[i]
data_p2 = pd.read_excel("Dane.xlsx", usecols="D,F")
array_p2 = data_p2.to_numpy()
```

Ze względu na to, że bezrobotni są destymuantą ich wskaźnik musi być ujemny

```python
for i in range(number_of_rows):
  array_p2[i, 1] = array_p2[i, 1]*(-1)
```

Tworzenie tablicy wskaźników cząstkowych unormowanych

```python
rows = [['Śląskie'], ['Dolnośląskie'], ['Mazowieckie'], ['Małopolskie'], ['Wielkopolskie'], ['Zachodniopomorskie'], ['Łódzkie'], ['Pomorskie'],
        ['Opolskie'], ['Kujawsko-pomorskie'], ['Lubuskue'], ['Podkarpackie'], ['Świętokrzyskie'], ['Lubelskie'], ['Warmińsko-mazurskie'], ['Podlaskie']]
prepared_data = np.concatenate((normalized_array, array_p2), axis=1)
named_data = np.concatenate((rows, prepared_data), axis=1)
indicators = pd.DataFrame(named_data, columns=['Województwo', 'Wskaźnik małżeństw', 'Wskaźnik kultury',
                          'Wskaźnik gruntów leśnych', 'Wskaźnik dróg', 'Wskaźnik B+R', 'Wskaźnik pracujących', 'Wskaźnik bezrobotnych'])
indicators
```

Wyliczenie wskaźnika syntetycznego będącego średnią arytmetyczną unormowanych wskaźników cząstkowych

```python
mean_=np.zeros(16)
sum_=np.sum(prepared_data,axis=1).tolist()
for i in range(16):
  mean_[i]=sum_[i]/7
rows = ['Śląskie', 'Dolnośląskie', 'Mazowieckie', 'Małopolskie','Wielkopolskie', 'Zachodniopomorskie', 'Łódzkie', 'Pomorskie',
      'Opolskie', 'Kujawsko-pomorskie','Lubuskue', 'Podkarpackie', 'Świętokrzyskie', 'Lubelskie', 'Warmińsko-mazurskie', 'Podlaskie']
mean_indicator=np.column_stack((rows,mean_))
mean_indicators = pd.DataFrame(mean_indicator,columns=['Województwo','Wskaźnik syntetyczny'])
mean_indicators
```

Przeprowadzenie rankingu na podstawie wskaźnika syntetycznego oraz posortowanie województw względem rankingu

```python
mean_indicators['Rank']=mean_indicators['Wskaźnik syntetyczny'].rank(ascending=False)
sorted=mean_indicators.sort_values(by=['Rank'])
sorted
```
